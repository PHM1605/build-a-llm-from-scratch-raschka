 ## Simplified self-attention

```python
# (max_length,dim) assuming batch_size=1, with max_length=sentence length
import torch
inputs = torch.tensor(
    [[0.43, 0.15, 0.89], # Your
    [0.55, 0.87, 0.66], # journey
    [0.57, 0.85, 0.64], # starts
    [0.22, 0.58, 0.33], # with
    [0.77, 0.25, 0.10], # one
    [0.05, 0.80, 0.55]] # step
)
```

```python
query = inputs[1]
attn_scores_2 = torch.empty(inputs.shape[0]) # (max_length,)
for i, x_i in enumerate(inputs):
    attn_scores_2[i] = torch.dot(x_i, query)
print(attn_scores_2)
```

```python
# normalized attention scores for each word
attn_weights_2_tmp = attn_scores_2 / attn_scores_2.sum()
print("Attention weight:", attn_weights_2_tmp)
print("Sum:", attn_weights_2_tmp.sum())
```

```python
# normalized attention scores based on <softmax>, naively
def softmax_naive(x):
    return torch.exp(x) / torch.exp(x).sum(dim=0)
attn_weights_2_naive = softmax_naive(attn_scores_2)
print("Attention weights:", attn_weights_2_naive)
print("Sum:", attn_weights_2_naive.sum())
```

```python
# normalized attention scores based on <softmax> from torch
attn_weights_2 = torch.softmax(attn_scores_2, dim=0)
print("Attention weights:", attn_weights_2)
print("Sum:", attn_weights_2.sum())
```

```python
# 2nd context vector z[2]
query = inputs[1] # (3,) - 2nd word token i.e. [0.55,0.87,0.66]
context_vec_2 = torch.zeros(query.shape) # return vector of dimension (3,)
for i,x_i in enumerate(inputs):
    context_vec_2 += attn_weights_2[i]*x_i
print(context_vec_2)
```

### Attention weights for all input tokens (instead of only 2nd element like above)

```python
attn_scores = torch.empty(6,6)
for i, x_i in enumerate(inputs):
    for j, x_j in enumerate(inputs):
        attn_scores[i, j] = torch.dot(x_i, x_j)
print(attn_scores)
```

```python
# attention_scores calculated with matrix multiplication
attn_scores = inputs @ inputs.T
print(attn_scores)
```

```python
# normalize each row => <attention_weights>
attn_weights = torch.softmax(attn_scores, dim=-1) # (6,6)
print(attn_weights)
```

```python
# verify
row_2_sum = sum([0.1385, 0.2379, 0.2333, 0.1240, 0.1082, 0.1581])
print("Row 2 sum:", row_2_sum)
print("All row sums:", attn_weights.sum(dim=-1))
```

```python
# 1st row of <attn_weights> will scale all inputs then add up => 1st context row
all_context_vecs = attn_weights @ inputs
print(all_context_vecs)
# verify
print("Previous 2nd context vector:", context_vec_2)
```

## Self-attention with trainable weights

```python
# Quick test for 1 sample (3,) first
x_2 = inputs[1]
d_in = inputs.shape[1]
d_out = 2

torch.manual_seed(123)
W_query = torch.nn.Parameter(torch.rand(d_in, d_out), requires_grad=False)
W_key = torch.nn.Parameter(torch.rand(d_in, d_out), requires_grad=False)
W_value = torch.nn.Parameter(torch.rand(d_in, d_out), requires_grad=False)

query_2 = x_2 @ W_query # (2,)
key_2 = x_2 @ W_key
value_2 = x_2 @ W_value
print(query_2) # (1,2)
```

```python
# for all samples
keys = inputs @ W_key # (6,2)
values = inputs @ W_value # (6,2)
print("keys.shape:", keys.shape)
print("values.shape:", values.shape)
```

```python
# calculate w_22 (attention score) with query is word-1
keys_2 = keys[1]
attn_scores_22 = query_2.dot(keys_2)
print(attn_scores_22)
```

```python
# all attention scores when query is word-1
attn_scores_2 = query_2 @ keys.T # (1,2) @ (2,6) -> (1,6)
print(attn_scores_2) # (1,6)
```

```python
# scaling & normalizing => <attention_weights> with 2nd-word-query (1,6)
d_k = keys.shape[-1] # 2
attn_weights_2 = torch.softmax(attn_scores_2 / d_k**0.5, dim=-1)
print(attn_weights_2)
```

```python
# context vector with 2nd-word-query
# is weighted-sum of <value-vector>
context_vec_2 = attn_weights_2 @ values # (1,6) @ (6,2) -> (1,2)
print(context_vec_2)
```

## Compact self-attention class

```python
import torch.nn as nn

class SelfAttention_v1(nn.Module):
    def __init__(self, d_in, d_out):
        super().__init__()
        self.W_query = nn.Parameter(torch.rand(d_in, d_out)) # (3,2)
        self.W_key = nn.Parameter(torch.rand(d_in, d_out)) # (3,2)
        self.W_value = nn.Parameter(torch.rand(d_in, d_out)) # (3,2)
        
    def forward(self, x):
        keys = x @ self.W_key # (batch,2)
        queries = x @ self.W_query # (batch,2)
        values = x @ self.W_value # (batch,2)
        # each row is one word; comparing that word with each word
        attn_scores = queries @ keys.T # (batch,2)@(2,batch)=>(batch,batch)
        # normalize each row, (batch,batch)
        attn_weights = torch.softmax(
            attn_scores/keys.shape[-1]**0.5, dim=-1
        )
        # 1st row: when we choose 1st word as query
        # 2nd row: when we choose 2nd...
        context_vec = attn_weights @ values # (batch,batch)@(batch,2)=>(batch,2)
        return context_vec
    
```

```python
torch.manual_seed(123)
sa_v1 = SelfAttention_v1(d_in, d_out)
print(sa_v1(inputs))
```

## Self-attention with Linear torch layer

```python
class SelfAttention_v2(nn.Module):
    def __init__(self, d_in, d_out, qkv_bias=False):
        super().__init__()
        self.W_query = nn.Linear(d_in, d_out, bias=qkv_bias)
        self.W_key = nn.Linear(d_in, d_out, bias=qkv_bias)
        self.W_value = nn.Linear(d_in, d_out, bias=qkv_bias)
    
    def forward(self, x):
        keys = self.W_key(x) # (batch,2)
        queries = self.W_query(x) # (batch,2)
        values = self.W_value(x) # (batch,2)
        attn_scores = queries @ keys.T # (batch,batch)
        # normalize each row; (batch,batch)
        attn_weights = torch.softmax(
            attn_scores / keys.shape[-1]**0.5, dim=-1
        )
        context_vec = attn_weights @ values # (batch,2)
        return context_vec 
        
torch.manual_seed(789)
sa_v2 = SelfAttention_v2(d_in, d_out)
print(sa_v2(inputs))
```
